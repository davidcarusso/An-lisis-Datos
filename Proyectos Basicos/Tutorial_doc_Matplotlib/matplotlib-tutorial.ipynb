```python
import matplotlib.pyplot as plt
```

# Ejemplo 1

plt.plot(x, y, label='label', color='color', linestyle='linestyle', marker='marker'): 
Traza una línea o un marcador en el gráfico.

```python
# Ejemplo 1

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, label="Grafico 1", color="blue", linestyle="-", marker="o")
plt.xlabel("Ejemplo x")
plt.ylabel("Ejemplo y")
plt.title("Grafico de ejemplo")
plt.legend()
plt.show()
```

# Ejemplo 2

plt.scatter(x, y, color='color', marker='marker'): 
Traza un gráfico de dispersión de puntos.

```python

plt.scatter(x, y, label="Dispersion de puntos", color="red")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Grafico de ejemplo Nº 2")
plt.legend()

plt.show()

```

# Ejemplo 3

plt.bar(x, height, width=0.8, color='color'): 
Traza un gráfico de barras verticales.


```python

x = ['A', 'B', 'C', 'D', 'E']
height = [10, 15, 7, 10, 12]

plt.bar(x, height, color="green")
plt.xlabel("Categorias")
plt.ylabel("Valor")
plt.title("Grafico de Barras")

plt.show()
```

# Ejemplo 4

plt.hist(x, bins, color='color')

Traza un histograma.

```python
import numpy as np
```

```python


#np.random.seed(10)         # fija un numero para que no sea aletorio los demas. (semilla)

data = np.random.randn(1000)      # trae numeros aletorios con coma

#print(data)

plt.hist(data, bins=30, label="Ejemplo 3", color='orange')     # desconozco que realiza el codigo bins=30
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.legend()
plt.show()


# Nota:         Averiguar para que sirve un histograma.
```
