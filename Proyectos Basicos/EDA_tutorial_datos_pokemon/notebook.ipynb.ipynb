```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ruta = ".\Pokemon.csv"
df = pd.read_csv(ruta)
df.head(15)
df.shape
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["type2"] = df["type2"].fillna(value="None")
df.isnull().sum()
```

```python
df = df.drop(["number"], axis=1)
df.info()
```

```python
tipo1_unicos = df.type1.unique()
sorted(tipo1_unicos)
```

```python
df["type1"] = df["type1"].replace("Graass", "Grass")
```

```python
tipo1_unicos = df.type1.unique()
sorted(tipo1_unicos)
```

```python
# vvalores unicas typo 2
sorted(df.type2.unique())
```

```python
# valores unicos legendarios

df["legendary"].unique()
```

Empezamos a graficar

```python


ax = ""

plt.figure(figsize=(12, 8))

sns.scatterplot(
    x=df["type1"],
    y=df["type2"],
    hue=df["legendary"]
)

plt.legend(loc="upper right")

plt.show()

# ax.tick_params(axis='x', rotation=90)
```

```python


plt.figure(figsize=(12, 8))

sns.countplot(
    data=df,
    x="type1",
    hue="legendary"
)

plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.xlabel('Type1')
plt.ylabel('Count')
plt.title('Count of Legendary Pokémon by Type1')
plt.legend(title='Legendary')

plt.show()
```
