```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

Vamos a intentar decifrar cual es el perfil de pasajero con mayor posivilidad de morir en Titanic.


Descripcion de los datos en dataset

  Column       Non-Null Count  Dtype  
---  ------       --------------  -----  
 0   PassengerId  418 non-null    int64     "Numero de pasajero o id"

 1   Survived     418 non-null    int64     "Si el pasajero sobrevivio o no (0 = Dead 1 = Alive)"

 2   Pclass       418 non-null    int64     "Clases 1°, 2° y 3° (1 = First class 2 = Second class 3 = Third class)"

 3   Name         418 non-null    object    "Nombre del pasajero"

 4   Sex          418 non-null    object    "Sexo del pasajero"

 5   Age          332 non-null    float64   "Edad del pasajero" 

 6   SibSp        418 non-null    int64     "Numero de acompañantes"

 7   Parch        418 non-null    int64  

 8   Ticket       418 non-null    object    "Numero de ticket"

 9   Fare         417 non-null    float64   "Tarifa del viaje"

 10  Cabin        91 non-null     object    "Nombre de la cabina"

 11  Embarked     418 non-null    object    "Lugar de embarque" (C = Cherbourg Q = Queenstown S = Southampton)
 

```python
# declaramos la ruta y cargamos el archivo csv
ruta = ".\dataset.csv"
df = pd.read_csv(ruta)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
# Completar la edad con promedio 
df["Age"] = df["Age"].fillna( df["Age"].mean())

# completar la tarifa con media

df["Fare"] = df["Fare"].fillna( df["Fare"].mean())

df.describe(include="all")
df.info()
```

```python
# Analizar las columnas categoricas irrelevantes

# Conteo de los niveles en las diferentes columnas categoricas:
cols_cat = ["Name", "Sex", "Ticket", "Cabin", "Embarked"]

for col in cols_cat:
    print(f"Columna {col}: { df[col].nunique() } subniveles.")



# con esto se borraria si alguna columna tiene un solo dato unico, porq no da margen a analisis.
print("-----------------------------")
```

```python
# Analizar las columnas numericas 
df.describe()

# Analizando la std (la variacion estandar) que no sea == 0  
```

<h1>Eliminar datos repetidos</h1>



```python
# Eliminar filas repetidas:

print(f"Tamaño del set antes de eliminar las filas repetidas {df.shape}")

df.drop_duplicates(inplace=True)

print(f"Tamaño del set despues de eliminar las filas repetidas {df.shape}")
```

Se observa que no tiene datos duplicados en el set de datos.

<h1> Analizar columnas numericas.</h1>
<p> Se crea una variante cols_num

```python
cols_num = ["PassengerId", "Survived", "Pclass", "Age", "SibSp", "Parch", "Fare" ]


fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8,30) )
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(cols_num):
    sns.boxplot( x=col, data=df, ax=ax[i] )
    ax[i].set_title(col)

plt.show()
```
